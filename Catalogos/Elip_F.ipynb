```python
import numpy as np
import matplotlib.pylab as plt
```

```python
r1 = np.loadtxt('b1r10.vae',delimiter=',')
r12 = np.loadtxt('b1r12.vae',delimiter=',')
r14 = np.loadtxt('b1r14.vae',delimiter=',')
r16 = np.loadtxt('b1r16.vae',delimiter=',')
```

```python
def datosGeo(Catalogo):
    a=Catalogo[:,6]
    b=Catalogo[:,7]
    c=Catalogo[:,8]
    def elip(a,b,c):
        return (1/4)*(a**2 - c**2)/(a**2 + b**2 + c**2)
    def elip1(a,c):
        return 1 - (c/a)
    def prol(a,b,c):
        return (1/4)*((b**2 - c**2)+ (b**2 - a**2))/(a**2 + b**2 + c**2)
    rt=[elip1(a,c),elip(a,b,c),prol(a,b,c)]
    
    return rt
```

```python
geo_1 = datosGeo(r1)
geo_12 = datosGeo(r12)
geo_14 = datosGeo(r14)
geo_16 = datosGeo(r16)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.hist(geo_1[0],bins=10,histtype='step',label='r = 1.0',density = True,color='black')
plt.hist(geo_12[0],bins=10,histtype='step',label='r = 1.2',density = True,linestyle='dashed',color='black')
plt.hist(geo_14[0],bins=10,histtype='step',label='r = 1.4',density = True,linestyle='dotted',color='black')
plt.hist(geo_16[0],bins=10,histtype='step',label='r = 1.6',density = True,linestyle='dashdot',color='black')
plt.legend(loc=0.0)
plt.xlabel(r'$\epsilon = 1 - c/a$')

plt.subplot(122)
plt.hist(geo_1[1],bins=10,histtype='step',label='r = 1.0',density = True,color='black')
plt.hist(geo_12[1],bins=10,histtype='step',label='r = 1.2',density = True,linestyle='dashed',color='black')
plt.hist(geo_14[1],bins=10,histtype='step',label='r = 1.4',density = True,linestyle='dotted',color='black')
plt.hist(geo_16[1],bins=10,histtype='step',label='r = 1.6',density = True,linestyle='dashdot',color='black')
plt.legend(loc=0.0)
plt.xlabel(r'$\epsilon = \frac{1}{4}$' + ' ' + r'$\frac{a^2 - c^2}{a^2 + b^2 + c^2}$')

plt.savefig('elip_hist_black.pdf')
```
